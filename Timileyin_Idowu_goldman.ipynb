```python
import pandas as pd # for generating dataframes and data exploration
from random import randint #helps to generate random numbers
```

```python
# MAIN CODE/FUNCTION
def get_true_false(data_frame, randomArray): #create a function to get boolean values
    output_array = [] # an empty array to store new dataframe from for loop
    for row in range(len(data_frame)): #loop through the rows of the dataframe
        temp = [] #a temporary array to store the boolean comparison for each row
        
        
        for col in range(len(data_frame.iloc[row])):
            if data_frame.iloc[row, col] > randomArray[row]: #if the value in the dataframe is greater than the random being looped through
                temp.append(True) #append true to the temporary array
            else: #address scenarios that dont meet the condition above
                temp.append(False)
        output_array.append(temp)

    
    return pd.DataFrame(data=output_array) # return a dataframe from the output_array
```

```python

# THE CODE BELOW IS FOR TESTING PURPOSES ONLY
random_list = [randint(0,10001) for i in range(10)] # create a list of 10 random numbers between 1 and 10000

random_dataframe = [] # create an empty list to store the random numbers from the for loop below

for i in range(10): #loop through 0 and 9
    random_dataframe.append([randint(0,10000) for i in range(500)]) #create a list of 500 random numbers and append to random_dataframe

df = pd.DataFrame(data=random_dataframe) #create a dataframe from the random_dataframe

print(random_list)
df
```
