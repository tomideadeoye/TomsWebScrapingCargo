```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://africacheck.org/'
pre_soup = BeautifulSoup(urlopen(Request(url, headers={'User-Agent' :  'Mozilla/5.0'})).read(), 'html.parser')
```

```python
needed_soup = pre_soup.find_all("article")
```

```python
needed_soup
final_data = []



for i in needed_soup[1:8]: # ignoring the first article which is just a banner
    try: 
        lines = list(filter(None, i.find('h3' < 'a').text.strip().split("\n")))
        final_data.append({
            'link':  url + i.find('a').get('href'),
            'main_text': lines[-3],
            'sub_text': lines[-2],
            'date': lines[-1]
        })
    except:
        pass

for i in final_data:
    print(i)
```
