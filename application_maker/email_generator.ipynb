```python
def getPerson(target_company , role, name):
    try:
        google_response = google_search("site:linkedin.com/in intitle:"+role+" "+ target_company['company_name'])
        if google_response == None:
            return 

        target_linkedin_person = google_response[0]
        linkedinPersonName = target_linkedin_person['title'].split(" - ")[0].split(" ") # ['John', 'Doe']
        first_name, last_name = linkedinPersonName[0], linkedinPersonName[1] # John Doe

        print(linkedinPersonName, role, target_linkedin_person['link'], target_linkedin_person['snippet'])

        if name['first_name'] != None:
            first_name = name['first_name']
        if name['last_name'] != None:
            last_name = name['last_name']

        valid_email = emailGenerator(first_name, last_name, target_company['company_domain']) # generates a list of possible email addresses:
        print({
                "name": linkedinPersonName,
                "role": role,
                "company_name": target_company['company_name'],
                "company_domain": target_company['company_domain'],
                "linkedin_url": target_linkedin_person['link'],
                "person_snippet": target_linkedin_person['snippet'],
                "company_snippet": target_company['snippet'],

                "email": valid_email
            }) 

        return {
                "name": linkedinPersonName,
                "role": role,
                "company_name": target_company['company_name'],
                "company_domain": target_company['company_domain'],
                "linkedin_url": target_linkedin_person['link'],
                "person_snippet": target_linkedin_person['snippet'],
                "company_snippet": target_company['snippet'],
                "email": valid_email
            }
            
    except Exception as e:
        print(e)
        pass
```

```python
target_company_list = ["Insight Partners", "QED Investors", "Harlem Capital", ]
target_company_list = ["Manta Network"]

back_up_list = []

def email_aggregator(company_name):
    try:
        google_response = google_search(company_name + " company")
        if google_response == None:
            return 

        companies= [{
                "company_name": company_name,
                "title": company['title'],
                "link": company['link'],
                "snippet": company['snippet'],
                "company_domain": company['displayLink'],
            } for company in google_response]

        all_company_details =  [detail['snippet'] for detail in companies] 

        companies = [company for company in companies if "linkedin" not in company['link'] or "crunchbase" not in company['link']]

        target_company = companies[0]
        
        target_company['details'] = all_company_details

        name = {"first_name": None,  "last_name": None}
        response = [getPerson(target_company, role,  name) for role in ['ceo', 'founder', 'recruiter', 'partner']] # exceute function getPerson for each role in the list
        back_up_list.append(response)
        return response

    except Exception as e:
        print(e)
        pass

draft_list = [email_aggregator(company) for company in target_company_list]




```
