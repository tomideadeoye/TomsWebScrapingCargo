```python
source_dir = "/Users/tomideisawesome/Downloads"
dest_dir_documents = "/Users/tomideisawesome/Documents/GitHub/TomsWebScrapingCargo/pluralsight_files/pdf"
dest_dir_dart = "/Users/tomideisawesome/Documents/GitHub/TomsWebScrapingCargo/pluralsight_files/msg"
document_extensions = []
dart_extensions = [".msg"]
# document_extensions = [".doc", ".docx", ".odt", ".pdf", ".xls", ".xlsx", ".ppt", ".pptx"]
# image_extensions = [".jpg", ".jpeg", ".jpe", ".jif", ".jfif", ".jfi", ".png", ".gif", ".webp", ".tiff", ".tif", ".psd", ".raw", ".arw", ".cr2", ".nrw", ".k25", ".bmp", ".dib", ".heif", ".heic", ".ind", ".indd", ".indt", ".jp2", ".j2k", ".jpf", ".jpf", ".jpx", ".jpm", ".mj2", ".svg", ".svgz", ".ai", ".eps", ".ico"]
# video_extensions = [".webm", ".mpg", ".mp2", ".mpeg", ".mpe", ".mpv", ".ogg", ".mp4", ".mp4v", ".m4v", ".avi", ".wmv", ".mov", ".qt", ".flv", ".swf", ".avchd"]
# audio_extensions = [".m4a", ".flac", "mp3", ".wav", ".wma", ".aac"]
# dest_dir_sfx = ""
# dest_dir_music = ""
# dest_dir_video = ""
# dest_dir_image = ""
```

```python
def make_unique(dest, name):
    filename, extension = splitext(name)
    counter = 1
    # * IF FILE EXISTS, ADDS NUMBER TO THE END OF THE FILENAME
    while exists(f"{dest}/{name}"):
        name = f"{filename}({str(counter)}){extension}"
        counter += 1

    return name

def move_file(dest, entry, name):
    if exists(f"{dest}/{name}"):
        unique_name = make_unique(dest, name)
        oldName = join(dest, name)
        newName = join(dest, unique_name)
        rename(oldName, newName)
    move(entry, dest)



def check_document_files(entry, name):  # * Checks all Document Files
    for documents_extension in document_extensions:
        if name.endswith(documents_extension) or name.endswith(documents_extension.upper()):
            move_file(dest_dir_documents, entry, name)
            logging.info(f"Moved document file: {name}")

def check_dart_files(entry, name):  # * Checks all dart Files
    for dart_extension in dart_extensions:
        if name.endswith(dart_extension) or name.endswith(dart_extension.upper()):
            move_file(dest_dir_dart, entry, name)
            logging.info(f"Moved dart file: {name}")



def check_files(source_dir):
    obj = os.scandir(source_dir)
    file_counter = 0
    folder_counter = 0
    for entry in obj:
        if entry.is_file():
            file_counter += 1
            check_document_files(entry, entry.name)
            check_dart_files(entry, entry.name)
        elif entry.is_dir():
            print("directory", entry.name)
            folder_counter += 1
            # print(entry.name)
            check_files(entry.path)

    print(file_counter, "files found")
    print(folder_counter, "folders found")

check_files(source_dir)

```
