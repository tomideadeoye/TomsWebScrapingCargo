**Import Dependencies**

```python
from requests import get
from bs4 import BeautifulSoup as Soup
import pandas as pd
```

```python
master_list = []
```

```python
master_list = []
def find_hightest(starting_url):
    response = get(starting_url)
    Soup_data = Soup(response.text, 'html.parser')
    Movies = Soup_data.findAll("div",{'class':'lister-item-content'})
    the_most_high = []
    the_most_high_counter = 0
    for movie in Movies:
        try:
            runtime = int(movie.find("span",{'class':'runtime'}).text.split(" ")[0])
            if runtime > the_most_high_counter:
                the_most_high = [movie.find("h3",{'class':'lister-item-header'}).find("a").text, runtime]
                the_most_high_counter = runtime
                print(the_most_high)
        except Exception as e:
            pass
    master_list.append(the_most_high)
    
```

```python
for i in range(1000, 8001, 50):
    try:
        starting_url = "https://www.imdb.com/search/title/?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&start=" + str(i) + "&ref_=adv_nxt"
        find_hightest(starting_url)
        print(i)
    except Exception as e:
        print("error", "https://www.imdb.com/search/title/?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&start=" + str(i) + "&ref_=adv_nxt")
        print(e)
```

```python
master_list_df = pd.DataFrame(master_list, columns = ['Movie', 'Runtime'])
master_list_df.sort_values(by=["Runtime"], ascending=False)
```
